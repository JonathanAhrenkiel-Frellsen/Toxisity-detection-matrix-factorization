```python
import pandas as pd
import numpy as np
```

```python
comments_info = pd.read_csv("./data/toxicity_annotated_comments.tsv", sep='\t')
comments_info.set_index("rev_id", inplace=True)
comments_people_bridge = pd.read_csv("./data/toxicity_annotations.tsv", sep='\t')
people_info = pd.read_csv("./data/toxicity_worker_demographics.tsv", sep='\t')
```

```python
def get_comment_from_id (rev_id):
    return comments_info.loc[rev_id]

def to_readalbe_text (text):
    return text.replace("NEWLINE_TOKEN", "\n")
```

```python
comments_people_bridge_var = comments_people_bridge.groupby(['rev_id']).var()
```

```python
comments_people_bridge_mean = comments_people_bridge.groupby(['rev_id']).mean()

comments_people_bridge_mean.where(comments_people_bridge_mean["toxicity_score"] < 0).dropna()
```
